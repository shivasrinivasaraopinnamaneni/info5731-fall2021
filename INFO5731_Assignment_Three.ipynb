```python
import pandas as pd
df = pd.read_csv("/content/reviews.csv")
df.head()
```

```python
!pip install nltk
import nltk
#Remove noise
df['review_text'] = df['review_text'].str.replace('[^\w\s]','')
#df['tweet_text'].head()

#Removing stopwords
nltk.download('stopwords')
from nltk.corpus import stopwords
stop = stopwords.words('english')
df['review_text'] = df['review_text'].apply(lambda x: " ".join(x for x in str(x).split() if x not in stop))
df['review_text']

#Removing numbers
df['review_text'] = df['review_text'].apply(lambda x: " ".join(x for x in x.split() if not x.isdigit()))
df['review_text']

#converting to lowercase
df['review_text'] = df['review_text'].apply(lambda x: " ".join(x.lower() for x in x.split()))
df['review_text']

#stemming
from nltk.stem import PorterStemmer
st = PorterStemmer()
df['review_text'][:5].apply(lambda x: " ".join([st.stem(word) for word in x.split()]))

#lemmatization
!pip install -U textblob
from textblob import Word
import nltk
nltk.download('wordnet')

df['review_text'] = df['review_text'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
df['review_text'].head()
```

# **Question 2: Undersand TF-IDF and Document representation**

(40 points). Starting from the documents (all the reviews, or abstracts, or tweets) collected for assignment two, write a python program: 

(1) To build the **documents-terms weights (tf*idf) matrix bold text**.

(2) To rank the documents with respect to query (design a query by yourself, for example, "An Outstanding movie with a haunting performance and best character development") by using **cosine similarity**.

```python
#To build the documents-terms weights (tf*idf) matrix bold text.
#calculating tf 
tf_result = (df['review_text']).apply(lambda x: pd.value_counts(x.split(" "))).sum(axis = 0).reset_index()
tf_result.columns = ['review_text','tf']
import numpy as np
#calculating tf*idf
for i,j in enumerate(tf_result['review_text']):
  tf_result.loc[i, 'idf'] = np.log(df.shape[0]/(len(df[df['review_text'].str.contains(j)])))
tf_result['tf*idf'] = tf_result['tf'] * tf_result['idf']
tf_result 



```

```python
#Rank the documents with respect to query by using cosine similarity.
#importing libraries
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from nltk.corpus import stopwords
import numpy as np
import numpy.linalg as LA
```

```python
tn_set = df['review_text'].values.tolist() 
tt_set = "iphone 11 which is unlocked looks like brand new and also performs equal to new phone" 
tt_set = [tt_set]
stopWords = stopwords.words('english')
vzer = CountVectorizer(stop_words = stopWords)
transformer = TfidfTransformer()
print(transformer)
```

```python
tnVectorizerArray = vzer.fit_transform(tn_set).toarray()
ttVectorizerArray = vzer.transform(tt_set).toarray()
cx = lambda a, b : np.inner(a, b)/(LA.norm(a)*LA.norm(b))
cosine_values = []
for vector in tnVectorizerArray:
        for testV in ttVectorizerArray:
            cosine = cx(vector, testV)
            cosine_values.append(cosine)
df_result = df.filter(['body '], axis=1)
se = pd.Series(cosine_values)
df_result['Cosine_similarity'] = se.values
df_result
```
